```python
import pandas as pd

df = pd.read_csv("weak-results.csv")
```

```python
total = df.groupby("Type")["Result"].count()
successes = df.groupby("Type")["Result"].apply(lambda x: (x == "Success").sum())

stats = pd.DataFrame({
    "Successes": successes,
    "Total": total,
})
stats["Success Rate"] = (stats["Successes"] / stats["Total"] * 100).round(1)

print("Attack Success by Prompt Type\n")
print(stats.to_string())

print("\nOverall: {}/{} attacks succeeded".format((df["Result"] == "Success").sum(), len(df)))
```
